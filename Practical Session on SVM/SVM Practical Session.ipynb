```python
import numpy as np
X = np.array([[-1, -1], [-2, -1], [1, 1], [2, 1]]) #we create 4 examples
y = np.array([-1, -1, 1, 1])
from sklearn.svm import SVC #import de la classe SVC pour SVM
classif=SVC() #we create a SVM with default parameters
classif.fit(X,y) #we learn the model according to given data
res=classif.predict([[-0.8, -1]]) #prediction on a new sample
print(res)
```

```python
import matplotlib.pyplot as plt #the library for plotting
#we create a mesh to plot in
h = .02 # grid step
x_min= X[:, 0].min() - 1
x_max= X[:, 0].max() + 1
y_min = X[:, 1].min() - 1
y_max = X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
np.arange(y_min, y_max, h))
print(xx)
print(yy)
#the grid is created, the intersections are in xx and yy
```

```python
mysvc= SVC(kernel='linear', C = 2.0)
mysvc.fit(X,y)
print(mysvc.score(X,y))
Z2d = mysvc.predict(np.c_[xx.ravel(),yy.ravel()]) # we predict all the grid

Z2d=Z2d.reshape(xx.shape)

plt.figure()
plt.pcolormesh(xx,yy,Z2d, cmap=plt.cm.Paired)
# We plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)
plt.show()
mysvc.support_vectors_
```

```python
#Non Linear Case 
import numpy as np
from sklearn.datasets import make_classification
X1, Y1 = make_classification(n_features=2, n_redundant=0, n_informative=2,
                             n_clusters_per_class=1)
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1,
            s=25, edgecolor='k')
plt.show()
```

Generating Points using Make Classification

```python
SVMC1= SVC()
#print(SVMC1)
SVMC1.fit(X1,Y1)
print(SVMC1.score(X1,Y1))
h = 0.02 # grid step
x_min= X1[:, 0].min() - 1
x_max= X1[:, 0].max() + 1
y_min = X1[:, 1].min() - 1
y_max = X1[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
np.arange(y_min, y_max, h))
Z2 = SVMC1.predict(np.c_[xx.ravel(),yy.ravel()])
Z2=Z2.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx,yy,Z2, cmap=plt.cm.Paired)
# We plot also the training points
plt.scatter(X1[:, 0], X1[:, 1], c=Y1, cmap=plt.cm.coolwarm)
plt.show()
```

The normal setting of SVM learns the model really good.

```python

import matplotlib.pyplot as plt #the library for plotting
#we create a mesh to plot in
h = 0.02 # grid step
x_min= X1[:, 0].min() - 1
x_max= X1[:, 0].max() + 1
y_min = X1[:, 1].min() - 1
y_max = X1[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
np.arange(y_min, y_max, h))
SVMC2= SVC(kernel='linear')
SVMC2.fit(X1,Y1)
print(SVMC2.score(X1,Y1))

Z2 = SVMC2.predict(np.c_[xx.ravel(),yy.ravel()])
Z2=Z2.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx,yy,Z2, cmap=plt.cm.Paired)
# We plot also the training points
plt.scatter(X1[:, 0], X1[:, 1], c=Y1, cmap=plt.cm.coolwarm)
plt.show()
```

When we use linear kernel it is not able to perform well.

```python
SVMC3= SVC(kernel='rbf', gamma= 0.06)
SVMC3.fit(X1,Y1)
print(SVMC3.score(X1,Y1))
Z2 = SVMC3.predict(np.c_[xx.ravel(),yy.ravel()])
Z2=Z2.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx,yy,Z2, cmap=plt.cm.Paired)
# We plot also the training points
plt.scatter(X1[:, 0], X1[:, 1], c=Y1, cmap=plt.cm.coolwarm)
plt.show()
```

```python
SVMC3= SVC(kernel='poly', degree=5)
SVMC3.fit(X1,Y1)
print(SVMC3.score(X1,Y1))
Z2 = SVMC3.predict(np.c_[xx.ravel(),yy.ravel()])
Z2=Z2.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx,yy,Z2, cmap=plt.cm.Paired)
# We plot also the training points
plt.scatter(X1[:, 0], X1[:, 1], c=Y1, cmap=plt.cm.coolwarm)
plt.show()
```

```python
a=[]
m= np.arange(1,100,1)
for i in range(1,100):
    SVMC3= SVC(kernel='rbf', gamma=i)
    SVMC3.fit(X1,Y1)
    a.append(SVMC3.score(X1,Y1))
    
plt.plot(m,a)
g= 'Gamma value for the RBF Kernel'
y= 'Score'
plt.xlabel(g)
plt.ylabel(y)
plt.show()
```

As, value of gamma increases, the performance of the RBF kernel increases the score of SVM.

# 3. Dataset Generation

#### 3.1 Random Dataset

```python
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=50,n_features=2, n_redundant=0, n_informative=2,
random_state=2, n_clusters_per_class=1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)
plt.show()
```

#### Task1: Tuning the parameters to find the best value of hyperparameter Gamma for RBF Kernel and Polynomial, through Cross-Validation.

```python
from sklearn.model_selection import cross_val_score
scores=[]
for i in range(1,50):
    score=0
    SVMC4= SVC(kernel= 'rbf', gamma= i)
    a= cross_val_score(SVMC4,X,y, cv=6)
    scores.append(score+(a[0]+a[1]+a[2]))
    
scores.append(score)
print('The best value of Gamma is:', scores.index(max(scores))+1 )
```

```python
from sklearn.model_selection import cross_val_score
scores=[]
for i in range(1,20):
    score=0
    SVMC4= SVC(kernel= 'poly', degree= i)
    a= cross_val_score(SVMC4,X,y,cv=6)
    scores.append(score+(a[0]+a[1]+a[2]))
 
scores.append(score)

print('The best value of Degree is:', scores.index(max(scores))+1 )
```

```python
from sklearn.model_selection import cross_val_score
scores=[]
for i in range(1,30):
    score=0
    SVMC4= SVC(kernel= 'poly', coef0= i)
    a= cross_val_score(SVMC4,X,y,cv=6)
    scores.append(score+(a[0]+a[1]+a[2]))
 
scores.append(score)

print('The best value of Bias is:', scores.index(max(scores))+1 )
```

#### Task 2: Plotting the decision boundaries and displaying support vectors.

Plotting for RBF Kernel with best Gamma found in previous task.

```python
import matplotlib.pyplot as plt #the library for plotting
#we create a mesh to plot in
h = 0.02 # grid step
x_min= X[:, 0].min() - 1
x_max= X[:, 0].max() + 1
y_min = X[:, 1].min() - 1
y_max = X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
np.arange(y_min, y_max, h))
SVMC2= SVC(kernel='rbf', gamma=9)
SVMC2.fit(X,y)
print('Score:',SVMC2.score(X,y))
a= SVMC2.support_vectors_
print('Number of Support Vectors are:', len(a))
Z2 = SVMC2.predict(np.c_[xx.ravel(),yy.ravel()])
Z2=Z2.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx,yy,Z2, cmap=plt.cm.Paired)
# We plot also the training points
plt.scatter(a[:, 0], a[:, 1], cmap=plt.cm.coolwarm)
plt.show()
```

Plotting the polynomial kernel with best degree and bias found after cross-validation.

```python
SVMC5= SVC(kernel='poly', degree=3, coef0=2)
SVMC5.fit(X,y)
print('Score:',SVMC5.score(X,y))
a= SVMC5.support_vectors_
print('Number of Support Vectors are:', len(a))
Z2 = SVMC2.predict(np.c_[xx.ravel(),yy.ravel()])
Z2=Z2.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx,yy,Z2, cmap=plt.cm.Paired)
# We plot also the training points
plt.scatter(a[:, 0], a[:, 1], cmap=plt.cm.coolwarm)
plt.show()
```

```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
iris = load_iris()
data1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])

x= data1.iloc[:,2:4]
y=data1.iloc[:,4:5]
y = np.ravel(y)
plt.scatter(x.iloc[:,0],x.iloc[:,1], c=y)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
scores=[]
for i in range(1,50):
    score=0
    SVMC5= SVC(kernel= 'rbf', gamma= i, decision_function_shape= 'ovr')
    a= cross_val_score(SVMC5,x,y, cv=3)
    scores.append(score+(a[0]+a[1]+a[2]))
    
scores.append(score)
print('The best value of Gamma is:', scores.index(max(scores))+1 )


from sklearn.model_selection import cross_val_score
scores=[]
for i in range(1,10):
    score=0
    SVMC6= SVC(kernel= 'poly', degree= i, decision_function_shape='ovr')
    a= cross_val_score(SVMC6,x,y,cv=3)
    scores.append(score+(a[0]+a[1]+a[2]))
 
scores.append(score)

print('The best value of Degree is:', scores.index(max(scores))+1 )


from sklearn.model_selection import cross_val_score
scores=[]
for i in range(1,30):
    score=0
    SVMC7= SVC(kernel= 'poly', coef0= i, decision_function_shape='ovr')
    a= cross_val_score(SVMC7,x,y,cv=3)
    scores.append(score+(a[0]+a[1]+a[2]))
 
scores.append(score)

print('The best value of Bias is:', scores.index(max(scores))+1 )
```

```python
import matplotlib.pyplot as plt #the library for plotting
#we create a mesh to plot in
h = 0.02 # grid step
x_min= x.iloc[:, 0].min() - 1
x_max= x.iloc[:, 0].max() + 1
y_min = x.iloc[:, 1].min() - 1
y_max = x.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
np.arange(y_min, y_max, h))
SVM= SVC(kernel='rbf', gamma=1)
SVM.fit(x,y)
print('Score:',SVM.score(x,y))
a= SVM.support_vectors_
print('Number of Support Vectors are:', len(a))
Z2 = SVM.predict(np.c_[xx.ravel(),yy.ravel()])
Z2=Z2.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx,yy,Z2, cmap=plt.cm.Paired)
# We plot also the training points
plt.scatter(a[:, 0], a[:, 1], cmap=plt.cm.coolwarm)
plt.show()
```
